## 1. Environment Setup & Imports

Load environment variables suitable for the project and import necessary libraries.

```python
import os
import sys
from dotenv import load_dotenv

# Add the src directory to sys.path to allow importing the chatbot module
sys.path.append(os.path.abspath(os.path.join('../src')))

from chatbot import RAGChatbot

# Load environment variables
load_dotenv()
```

## 2. Initialize Chatbot

Initialize the RAGChatbot class which handles embeddings, Pinecone connection, and LLM setup.

```python
try:
    chatbot = RAGChatbot()
    print("Chatbot initialized successfully.")
except Exception as e:
    print(f"Error initializing chatbot: {e}")
```

## 3. Data Ingestion (Optional)

Uncomment and run the cell below if you need to ingest new data from the 'data' directory.

```python
chatbot.ingest_data(data_path="../data")
```

## 4. Testing the Pipeline

Test the chatbot with some queries.

```python
query = "The CNC machine motor does not start and the fuses keep blowing. What are the possible causes and how can I fix the problem?"
response = chatbot.query(query)
print("Answer: \n", response["answer"])
```
